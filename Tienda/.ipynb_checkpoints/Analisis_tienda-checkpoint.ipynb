```python
import pandas as pd
import numpy as np
from IPython.display import display as dp
import os
```

```python
base='documents'
ventas=pd.read_csv(os.path.join(base,'Ventas.csv'))
inventario=pd.read_csv(os.path.join(base,'Inventario.csv'))
productos=pd.read_csv(os.path.join(base,'Productos.csv'))
tiendas=pd.read_csv(os.path.join(base,'Tiendas.csv'))
```

```python
dp(ventas.head())
```

```python
dp(inventario.head())
```

```python
dp(productos.head())
```

```python
dp(tiendas.head())
```

```python
df_analisis = ventas.merge(inventario, on = 'ProductoID', how = 'left')
```

```python
dp(df_analisis.head())
```

```python
df_analisis = df_analisis.merge(tiendas, on = 'TiendaID', how = 'left')
```

```python
dp(df_analisis.head())
```

```python
df_analisis = df_analisis.merge(productos, on = ['ProductoID', 'PrecioUnitario'], how = 'left')
```

```python
dp(df_analisis.head())
```

```python
df_analisis.info()
```

```python
df_analisis['Fecha'] = pd.to_datetime(df_analisis['Fecha'])
```

```python
df_analisis.info()
```

```python
df = df_analisis.copy() 
```

```python
df['Mes'] = df['Fecha'].dt.month
df['Mes_Nombre'] = df['Fecha'].dt.month_name()
```

```python
dp(df.head())
```

```python
# Cual es la Venta Por Mes?
```

```python
ingresos_por_mes=df.groupby(['Mes','Mes_Nombre'])['Ingreso'].sum().reset_index().drop(columns='Mes')
ingresos_por_mes.columns=['Mes', 'Ingreso']
```

```python
dp(ingresos_por_mes)
```

```python
ingresos_por_mes=df.groupby(['Mes','Mes_Nombre', 'Canal'])['Ingreso'].sum().reset_index().drop(columns='Mes')
ingresos_por_mes.columns=['Mes', 'Canal', 'Ingreso']
```

```python
dp(ingresos_por_mes)
```

```python
# Cuales son los 10 productos  mas vendidos por unidades y por ingreso?
```

```python
p_vendidos_ingresos = df.groupby('NombreProducto')['Ingreso'].sum()
```

```python
dp(p_vendidos_ingresos)
```

```python
# Utilizo la funcion nlargest una funcion de python para tomar los mas vendidos de acuerdo al numero que le proporciones
```

```python
top_10 = p_vendidos_ingresos.nlargest(10)
```

```python
dp(top_10)
```

```python
# Otra forma de hacerlo y mas estetica es resteando y convertirlo en una tabla
```

```python
p_vendidos_ingresos = df.groupby('NombreProducto')['Ingreso'].sum().reset_index()
top_10 = p_vendidos_ingresos.nlargest(10, 'Ingreso')
```

```python
dp(top_10)
```

```python
# Cuales son los 10 productos  mas vendidos por Cantidad?
```

```python
p_vendidos_ingresos = df.groupby('NombreProducto')['Cantidad'].sum().reset_index()
top_10_can = p_vendidos_ingresos.nlargest(10, 'Cantidad')
```

```python
dp(top_10_can)
```

```python
# Que categorias tienen el mayor margen promedio? es el resultado residual entre precio unitario y costo unitario
```

```python
df['Margen'] = df['PrecioUnitario']-df['CostoUnitario']
```

```python
dp(df.head())
```

```python
# Ya con el margen calculado ahora lo agrupo y le saco el promedio de cada categoria
```

```python
mc = df.groupby('Categoría')['Margen'].mean().reset_index()
```

```python
dp(mc)
```

```python
# Que productos estan bajo inventario  (<50 Unidades) y aun se siguen vendiendo?
```

```python
redflag = df[(df['StockActual'] < 50) & (df['Ingreso'] > 0)]
```

```python
dp(redflag)
```

```python
# Ahora solo escojo los datos que quiero mostrar para mejor observacion, hago un drop del 'index'
```

```python
redflag = df[['NombreProducto', 'StockActual','Ingreso']].reset_index().drop(columns='index')
dp(redflag)
dp(redflag.head())
```

```python
# Donde estan ubicadas las tiendas con menor Facturacion mensual
```

```python
mf = df.groupby(['NombreTienda', 'Ciudad', 'Mes', 'Mes_Nombre'])['Ingreso'].sum().reset_index().drop(columns='Mes')
mf.columns = ['Tienda', 'Ciudad', 'Mes', 'Ingreso']
```

```python
dp(mf.head())
```
