```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.listdir()

os.listdir("datasets")


# Opcional para mejor visualización
sns.set(style="whitegrid")
%matplotlib inline

# Cargar el archivo
df = pd.read_csv("datasets/Crime_Data_from_2020_to_Present.csv")

# Ver las primeras filas
df.head()
```

```python
# Ver número de filas y columnas
print("Forma del dataset:", df.shape)

# Ver nombres de columnas
print("\nColumnas del dataset:")
print(df.columns.tolist())

# Ver tipos de datos y nulos
print("\nInformación general del dataset:")
print(df.info())

# Ver cantidad de valores nulos por columna
print("\nValores nulos por columna:")
print(df.isnull().sum().sort_values(ascending=False))
```

```python
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M %p', errors='coerce')

# Generar columnas derivadas
df['Year'] = df['DATE OCC'].dt.year
df['Month'] = df['DATE OCC'].dt.month
df['Day'] = df['DATE OCC'].dt.day
df['Weekday'] = df['DATE OCC'].dt.day_name()

# Verificar que funciona
df[['DATE OCC', 'Year', 'Month', 'Day', 'Weekday']].head()
```

```python
# Conteo de crímenes por año
df['Year'].value_counts().sort_index().plot(kind='bar', figsize=(10, 5))
plt.title('Cantidad de Crímenes por Año')
plt.xlabel('Año')
plt.ylabel('Número de crímenes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Orden personalizado de los días
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Conteo de crímenes por día de la semana
df['Weekday'].value_counts().reindex(days_order).plot(kind='bar', figsize=(10, 5))
plt.title('Cantidad de Crímenes por Día de la Semana')
plt.xlabel('Día')
plt.ylabel('Número de crímenes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['Crm Cd Desc'].value_counts().head(10).plot(kind='barh', figsize=(10, 6))
plt.title('Top 10 Tipos de Crímenes Más Comunes')
plt.xlabel('Número de crímenes')
plt.gca().invert_yaxis()  # Para que el más común aparezca arriba
plt.tight_layout()
plt.show()
```
