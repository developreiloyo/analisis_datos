```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar el conjunto de datos desde el archivo CSV
df = pd.read_csv("World University Rankings.csv")

# Mostrar las primeras 5 filas del DataFrame original para exploración inicial
print("Primeras 5 filas del dataset original:")
print(df.head())
print("\n" + "-"*50 + "\n")
```

```python
# 2. Información general sobre el dataset
print("Tamaño del dataset: ", df.shape[0], "filas y", df.shape[1], "columnas")
print("Nombres de las columnas originales:")
print(df.columns.tolist())
print("\n" + "-"*50 + "\n")


# 3. Selección de columnas relevantes
# Selecciono solo las columnas que nos interesan para el análisis
columnas_relevantes = ['name', 'location', 'scores_teaching', 'scores_research', 'scores_citations']
df_limpio = df[columnas_relevantes]

# Renombro las columnas para facilitar su manejo y comprensión
df_limpio.columns = ['Universidad', 'Pais', 'Puntaje_Ensenanza', 'Puntaje_Investigacion', 'Citas']

# Eliminamos filas con valores nulos en la columna 'Puntaje_Investigacion'
# (esto me va a evitar errores al calcular promedios)
df_limpio = df_limpio.dropna(subset=['Puntaje_Investigacion'])

# Muestro los primeros registros del nuevo DataFrame limpio
print("Datos limpios (primeras 5 filas):")
print(df_limpio.head())
print("\n" + "-"*50 + "\n")
```

```python
# 4. Análisis estadístico: promedio de puntaje de investigación por país
# Agrupo por 'Pais' y calculamos el promedio del puntaje de investigación
promedio_investigacion = df_limpio.groupby('Pais')['Puntaje_Investigacion'].mean()

# Ordeno los países de mayor a menor puntaje promedio y seleccionamos los 10 primeros
top_paises_investigacion = promedio_investigacion.sort_values(ascending=False).head(10)

# Imprimo los resultados del top 10
print("Top 10 países por puntaje promedio de investigación:")
print(top_paises_investigacion)
print("\n" + "-"*50 + "\n")
```

```python
# 5. Visualización: gráfico de barras del top 10 países por investigación
plt.figure(figsize=(12, 6))
top_paises_investigacion.plot(kind='bar', color='skyblue')
plt.title('Top 10 Países por Puntaje Promedio de Investigación')
plt.xlabel('País')
plt.ylabel('Puntaje Promedio de Investigación')
plt.xticks(rotation=45, ha='right')  # Mejor visualización de etiquetas
plt.tight_layout()  # Ajusta automáticamente los márgenes
plt.show()
```
